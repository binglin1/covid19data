```python
import pandas as pd
```

```python
df_confirmed = pd.read_csv('time_series_covid19_confirmed_global_narrow.csv')
df_deaths    = pd.read_csv('time_series_covid19_deaths_global_narrow.csv')
df_recovered = pd.read_csv('time_series_covid19_recovered_global_narrow.csv')
```

```python
df_confirmed
```

```python
df_deaths
```

```python
df_recovered
```

```python
df_confirmed.drop(['ISO 3166-1 Alpha 3-Codes', 'Region Code', 'Sub-region Code','Intermediate Region Code'], axis=1, inplace=True)
df_confirmed = df_confirmed.rename(columns={'Value':'Confirmed'})

df_deaths.drop(['ISO 3166-1 Alpha 3-Codes', 'Region Code', 'Sub-region Code','Intermediate Region Code'], axis=1, inplace=True)
df_deaths = df_deaths.rename(columns={'Value':'Deaths'})

df_recovered.drop(['ISO 3166-1 Alpha 3-Codes', 'Region Code', 'Sub-region Code','Intermediate Region Code'], axis=1, inplace=True)
df_recovered = df_recovered.rename(columns={'Value':'Recovered'})
```

```python
df = pd.concat([df_confirmed, df_deaths, df_recovered], axis=1, join='outer')
```

```python
df = df.loc[:,~df.columns.duplicated()]
```
